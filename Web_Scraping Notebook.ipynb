```python
from bs4 import BeautifulSoup
import requests
```

```python
#The Weather Channel an IBM Business
html_text = requests.get("https://weather.com/weather/hourbyhour/l/Accra+Greater+Accra+Ghana?canonicalCityId=6ea38a5575bc43b7f51f1fd1416e23b944035f76b5e7f817354e76191c79a389").text
```

```python
soup = BeautifulSoup(html_text,'lxml')
```

```python
dayDetails = soup.find('details',class_="DaypartDetails--DayPartDetail--1up3g Disclosure--themeList--25Q0H Disclosure--disableBorder--24EOy")
hour = dayDetails.find('h2',class_="DetailsSummary--daypartName--2FBp2").text
precipitationPercentage = dayDetails.find('div',class_="DetailsSummary--precip--1ecIJ").text.replace('Rain','')
```

```python
hours = soup.findAll('h2',class_="DetailsSummary--daypartName--2FBp2")
temps = soup.findAll('span', class_="DetailsSummary--tempValue--1K4ka")
precipitationPercentages = soup.findAll('div',class_="DetailsSummary--precip--1ecIJ")
wind = soup.findAll('span', class_="Wind--windWrapper--3aqXJ undefined")
```

```python
temp = temps[0].text
print(temp)
```

```python
print(hours[1].text,temps[1].text,precipitationPercentages[1].text,wind[1].text)
```

```python
hourList = []
precipitationPercentagesList = []
for i in hours:
    hourList.append(i.text)
for i in precipitationPercentages:
    precipitationPercentagesList.append(i.text.replace('Rain',''))
for i in range(len(hourList)):
    print(hourList[i],precipitationPercentagesList[i])
```
